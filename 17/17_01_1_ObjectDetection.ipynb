```python
# (IoU ê³„ì‚° í•¨ìˆ˜) ë‘ ë°•ìŠ¤ì˜ êµì§‘í•©/í•©ì§‘í•© ë¹„ìœ¨
# IoU : ë‘ ë°•ìŠ¤ (boxA, boxB) ì–¼ë§ˆë‚˜ ê²¹ì¹˜ëŠ”ì§€ ë¹„ìœ¨ë¡œ ë‚˜íƒ€ë‚¸ ê°’ [0,1]
# 0 : ì•ˆ ê²¹ì³ìš” , 1: ì™„ì „ ê°™ì€ ìœ„ì¹˜
# box: [x1, y1, x2, y2]
# boxA = [10,10,60,60]
# boxB = [30,30,80,80]

def compute_iou(boxA, boxB):
  # ë‘ ë°•ìŠ¤ê°€ ê²¹ì³ ìˆì–´ìš”. ì™¼ìª½ ìœ„ (xA, yA), ì˜¤ë¥¸ìª½ ì•„ë˜(xB, yB)

  # ê²¹ì¹˜ëŠ” ì™¼ìª½ ìœ„ ì¢Œí‘œ
  xA = max(boxA[0], boxB[0]) # ê²¹ì¹˜ëŠ” ì˜ì—­ ì™¼ìª½ xì¢Œí‘œ
  # ì˜ˆì œ ì ìš©: max(10, 30) = 30
  yA = max(boxA[1], boxB[1]) # ê²¹ì¹˜ëŠ” ì˜ì—­ ì™¼ìª½ yì¢Œí‘œ
  # ì˜ˆì œ ì ìš©: max(10, 30) = 30

  # ê²¹ì¹˜ëŠ” ì˜¤ë¥¸ìª½ ì•„ë˜ ì¢Œí‘œ
  xB = min(boxA[2], boxB[2])  # ì˜ˆì œ ì ìš©: min(60, 80) = 60
  yB = min(boxA[3], boxB[3])  # ì˜ˆì œ ì ìš©: min(60, 80) = 60
  inter = max(0, xB - xA) *  max(0, yB - yA)
  # xB - xA : ê²¹ì¹˜ëŠ” ê°€ë¡œ ê¸¸ì´, yB - y ê²¹ì¹˜ëŠ” ì„¸ë¡œ ê¸¸ì´
  # ë‘ ê°’ ëª¨ë‘ 0 ë³´ë‹¤ ì‘ìœ¼ë©´ >> ê²¹ì¹˜ì§€ ì•Šì•„ìš”
  # êµì§‘í•© ë©´ì  : ê°€ë¡œ  60-30 = 30, ì„¸ë¡œ 60-30=30 >> 900

  areaA = (boxA[2] - boxA[0]) * ((boxA[3] - boxA[1])) # A ë©´ì 
  # 60-10 =50 * 50 = 2500

  areaB = (boxB[2] - boxB[0]) * ((boxB[3] - boxB[1])) # B ë©´ì 
  # 60-10 =50 * 50 = 2500

  union = areaA + areaB - inter   # í•©ì§‘í•©
  # 2500 + 2500 - 900 = 4100

  return inter / union if union >0 else 0
```

```python
import torch

# bbox iou = [x,y,w,h]

def iou_pytorch(boxes1, boxes2):
    """
    PyTorch ë²„ì „ IoU ê³„ì‚° í•¨ìˆ˜
    boxes1, boxes2: (..., 4) í˜•íƒœì˜ tensor
    box = [x1, y1, x2, y2]
    """

    # -----------------------------
    # 1. êµì§‘í•© ì˜ì—­ ê³„ì‚°
    # -----------------------------
    # ë‘ ë°•ìŠ¤ì˜ ì™¼ìª½ ìœ„ ì (x1, y1) ì¤‘ í° ê°’ ì„ íƒ â†’ êµì°¨ ì˜ì—­ì˜ ì‹œì‘ì 
    x1 = torch.max(boxes1[..., 0], boxes2[..., 0])
    y1 = torch.max(boxes1[..., 1], boxes2[..., 1])

    # boxes1[..., 0], boxes2[..., 0]
    # ..., : íŒŒì´ì¬ _ (ì•ˆ ê°€ì ¸ì˜¤ê² ë‹¤) ì˜ë¯¸ì™€ ìœ ì‚¬
    # >> ì• ìª½ì˜ ëª¨ë“  ì°¨ì› ê·¸ëŒ€ë¡œ ë‘ê³ , ë§ˆì§€ë§‰ ì°¨ì›ì—ì„œ ì¸ë±ìŠ¤ 0ë§Œ ì¶”ì¶œí•˜ê² ë‹¤ëŠ” ì˜ë¯¸

    # ë‘ ë°•ìŠ¤ì˜ ì˜¤ë¥¸ìª½ ì•„ë˜ ì (x2, y2) ì¤‘ ì‘ì€ ê°’ ì„ íƒ â†’ êµì°¨ ì˜ì—­ì˜ ëì 
    x2 = torch.min(boxes1[..., 2], boxes2[..., 2])  # w
    y2 = torch.min(boxes1[..., 3], boxes2[..., 3])  # h

    # êµì°¨ ì˜ì—­ì˜ width, height (ìŒìˆ˜ê°€ ë‚˜ì˜¤ë©´ 0ìœ¼ë¡œ ì²˜ë¦¬ â†’ ì•ˆ ê²¹ì¹˜ëŠ” ê²½ìš°)
    inter_w = torch.clamp(x2 - x1, min=0)
    inter_h = torch.clamp(y2 - y1, min=0)

    # êµì°¨ ì˜ì—­ ë©´ì 
    inter_area = inter_w * inter_h

    # -----------------------------
    # 2. ê° ë°•ìŠ¤ì˜ ë©´ì  ê³„ì‚°
    # -----------------------------
    area1 = (boxes1[..., 2] - boxes1[..., 0]) * (boxes1[..., 3] - boxes1[..., 1])
    area2 = (boxes2[..., 2] - boxes2[..., 0]) * (boxes2[..., 3] - boxes2[..., 1])

    # -----------------------------
    # 3. í•©ì§‘í•© ì˜ì—­
    # -----------------------------
    union = area1 + area2 - inter_area

    # 0 division ë°©ì§€
    eps = 1e-7

    # -----------------------------
    # 4. IoU ê³„ì‚°
    # -----------------------------
    iou = inter_area / (union + eps)

    return iou
```

```python
# ë‹¨ì¼ ë°•ìŠ¤ IoU
boxA = torch.tensor([10,10,60,60])
boxB = torch.tensor([30,30,80,80])

print(iou_pytorch(boxA, boxB))
# ì´ë¯¸ì§€ 1ê°œ >> ë°°ì¹˜ 1ê°œ
print(boxA.shape) # (4,)
```
