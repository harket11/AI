응용 예시

```python
# 두 변수를 비교해서 일치하는 개수로 정확도 계산

# 2개의 넘파이 배열 준비

# 정답
yt = np.array([1,1,0,1,0,1,1,0,1,0])

# 예측
yp = np.array([1,1,0,1,0,1,0,1,1,1])
```

```python
# 정답(y)과 예측(y_pred) 잘 맞나? (비교)
matched = (yt == yp)
print(matched)
```

```python
matched.sum()
```

```python
# 맞은 개수(정답 수 세는 방법)
correct = matched.sum()

# 전체 수
total = len(matched)
```

```python
# 정확도 계산
accuracy = correct / total
print(accuracy)
```

```python
print(f'정답수: {correct}, 전체수: {total}, 정확도:{accuracy*100}% ' )
```

```python
n1_max = n1.max()
n1_min = n1.min()
print(n1_max, n1_min)
```

```python
# 변환(브로드캐스팅 기능 사용해야지)
(n1-n1_min) / (n1_max -n1_min) # 정규화 (min-max scaling)
```
